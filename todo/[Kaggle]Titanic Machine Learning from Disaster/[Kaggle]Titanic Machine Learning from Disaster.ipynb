## Competition
* 這個題目是判斷鐵達尼號出事時, 每位乘客最後是存活 or 死亡
* 這是個二元分類問題 (Binary classification)

## Training data
* train.csv
* 各欄位說明
  - PassengerId: 乘客編號
  - Survived: 生(1), 死(0)
  - Pclass: 船票等級 1 = 1st(Upper), 2 = 2nd(Middle), 3 = 3rd(Lower)
  - Name: 姓名
  - Sex: 性別
  - Age: 年齡
  - SibSp: 共乘配偶和兄弟姊妹的數量
  - Parch: 共乘父母和子女的數量
  - Ticket: 船票號碼
  - Fare: 旅費
  - Cabin: 艙房號碼
  - Embarked: 登船港口

````python
import numpy as np
import pandas as pd
from keras import models
from keras import layers
from keras import optimizers
from keras import losses 
from keras import metrics
import matplotlib.pyplot as plt
````

## 載入 train and test dataset
* x_train: copy training_set and 移除 'PassengerId','Name','Ticket','Survived', 與存活與否可能無關的 data
* y_train: copy training_set 的 'Survived' 作為 labeling data

````python
training_set = pd.read_csv('./input/train.csv')
testing_set = pd.read_csv('./input/test.csv')

x_train = training_set.drop(['PassengerId','Name','Ticket','Survived'], axis=1)
y_train = training_set['Survived']
print("x_train.shape:%r" % (x_train.shape,))
print("y_train.shape:%r" % (y_train.shape,))
print("x_train:\n%r" % (x_train[:10]))

x_test = testing_set.drop(['PassengerId','Name','Ticket'], axis=1)
print("x_test.shape:%r" % (x_test.shape,))
````

* 將 "Age" 中 NaN value, 填入平均值取代

````python
x_train['Age'] = x_train['Age'].fillna(x_train['Age'].mean())
x_test['Age'] = x_test['Age'].fillna(x_test['Age'].mean())
print("x_train:\n%r" % (x_train[:10]))
````

````python
def simplify_ages(df):
    #df['Age'] = df['Age'].fillna(-0.5)
    bins = (-1, 0, 5, 12, 18, 25, 35, 60, 120)
    group_names = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student', 'Young Adult', 'Adult', 'Senior']
    categories = pd.cut(df['Age'], bins, labels=group_names)
    df['Age'] = categories.cat.codes 
    return df
````

* 0: 0~5 歲
* 1: 5~12
* 2: 12~18
* 3: 18~25
* 4: 25~35

````python
def simplify_cabins(df):
    df['Cabin'] = df['Cabin'].fillna('N')
    df['Cabin'] = df['Cabin'].apply(lambda x: x[0])
    df['Cabin'] =  pd.Categorical(df['Cabin'])
    df['Cabin'] = df['Cabin'].cat.codes 
    return df
````

* 簡化 "Cabin" 艙房號碼, 分別有 Axxx ~ Gxxx

* mapping tabel:

  - 7: N

````python
def simplify_fares(df):
    df['Fare'] = df.Fare.fillna(-0.5)
    bins = (-1, 0, 8, 15, 31, 1000)
    group_names = ['Unknown', '1_quartile', '2_quartile', '3_quartile', '4_quartile']
    categories = pd.cut(df['Fare'], bins, labels=group_names)
    df['Fare'] = categories.cat.codes 
    return df
````

* 1: 0~8

````python
def simplify_sex(df):
    df['Sex'] = pd.Categorical(df['Sex'])
    df['Sex'] = df['Sex'].cat.codes 
    return df
````

````python
def simplify_embarked(df):
    df['Embarked'] = pd.Categorical(df['Embarked'])
    df['Embarked'] = df['Embarked'].cat.codes + 1
    return df
````

````python
def transform_features(df, dump=False):
    df = simplify_ages(df)
    #if (dump == True): print("simplify_ages:\n%r" % (df[:10]))
    df = simplify_cabins(df)
    #if (dump == True): print("simplify_cabins:\n%r" % (df[:10]))
    df = simplify_fares(df)
    #if (dump == True): print("simplify_fares:\n%r" % (df[:10]))
    df = simplify_sex(df)
    #if (dump == True): print("simplify_sex:\n%r" % (df[:10]))
    df = simplify_embarked(df)
    #if (dump == True): print("simplify_embarked:\n%r" % (df[:10]))
    return df

transform_features(x_train)
transform_features(x_test)
````

## 建立 network(model)
* Layer 1: 全連接 
* Layer 2: 全連接
* Layer 3: 全連接
* Layer 4: 全連接, "sigmoid" output 介於 [0,1]

````python
model = models.Sequential()
model.add(layers.Dense(32, activation='relu', 
                       input_shape=(8,)))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(16, activation='relu'))
#model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.summary()

model.compile(optimizer=optimizers.RMSprop(lr=0.001),
             loss=losses.binary_crossentropy,
             metrics=[metrics.binary_accuracy])
````

* 將 train data x and y 分成兩組
* 前 691 筆為 train data
* 後 200 筆為 validation(test) data

````python
y_train = np.asarray(y_train)
x_train = np.asarray(x_train)
x_test = np.asarray(x_test)

validation_size = 200

x_val = x_train[:validation_size]
partial_x_train = x_train[validation_size:]
print("partial_x_train.shape:%r" % (partial_x_train.shape,))
print("x_val.shape:%r" % (x_val.shape,))

y_val = y_train[:validation_size]
partial_y_train = y_train[validation_size:]
````

## 開始訓練
* 可修改 epoch 觀察 loss and acc

````python
history = model.fit(partial_x_train, partial_y_train, epochs=30, validation_data=(x_val, y_val))
````

````python
acc = history.history['binary_accuracy']
val_acc = history.history['val_binary_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss,'b', label='Validation loss')

plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
````

* 看來有 overfitting 問題

````python
plt.clf()

epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'bo', label='Training Acc')
plt.plot(epochs, val_acc,'b', label='Validation Acc')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Acc')
plt.show()
````

````python
predictions = model.predict_classes(x_test)
ids = testing_set['PassengerId'].copy()
new_output = ids.to_frame()
new_output["Survived"]=predictions
new_output.head(10)
````
