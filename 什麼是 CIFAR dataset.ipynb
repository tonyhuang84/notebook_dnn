* The CIFAR-10 and CIFAR-100 are labeled subsets of the 80 million tiny images dataset.
  - 8千萬張 labeled 小圖(32 x 32 x 3)資料庫 (CIFAR-10 and CIFAR-100 各有 60000 images, 但 total 卻有 80 million images ???)
  - CIFAR-10: 10 個分類 dataset
  - CIFAR-100: 100 個分類 dataset
* They were collected by Alex Krizhevsky, Vinod Nair, and Geoffrey Hinton.

## The CIFAR-10 dataset
* The CIFAR-10 dataset consists of 60000 32x32 colour images in 10 classes, with 6000 images per class.
  - 10 個分類, 每個分類有 6000 圖片, 共 60000 張圖片
* There are 50000 training images and 10000 test images.
  - 已經預先設定好 training(50000) 和 test(10000) 圖片集
* The dataset is divided into five training batches and one test batch, each with 10000 images.
  - one class(6000) = training batches(5) * 1000 + test batch(1) * 1000
  - dataset(60000) = classes(10) * 6000

## The CIFAR-100 dataset
* This dataset is just like the CIFAR-10, except it has 100 classes containing 600 images each.
  - 100 個分類, 每個分類有 600 張圖片
* There are 500 training images and 100 testing images per class.
* The 100 classes in the CIFAR-100 are grouped into 20 superclasses.
* Each image comes with a "fine" label (the class to which it belongs) and a "coarse" label (the superclass to which it belongs).
