* Convolutional Neural Networks，CNN
* 1988 年，Yann LeCun, LeNet 是推进深度学习领域发展的最早的卷积神经网络之一

## ConvNet 有四个主要操作：
* 卷积
* 非线性处理（ReLU）
* 池化或者亚采样
* 分类（全连接层）



## 卷积 (Convolution)
* 目的是为了从输入图像中提取特征

* 考虑一下一个 5 x 5 的图像，它的像素值仅为 0 或者 1


* 同时，考虑下另一个 3 x 3 的矩阵



https://wx4.sinaimg.cn/mw690/67111f6aly1fhykes5sffg207g05gdgf.gif

* 3x3 的矩阵叫做滤波器（filter）or kernel
* 計算後輸出的矩陣稱為 feature map
* 不同的滤波器可以从图中检测到不同的特征，比如边缘、曲线等


### 特征图的大小（卷积特征）由下面三个参数控制
* 深度（Depth）：深度对应的是卷积操作所需的滤波器个数


* 步长（Stride）：步长是我们在输入矩阵上滑动滤波矩阵的像素数, 步长越大，将会得到更小的特征图
* 零填充（Zero-padding）: 在输入矩阵的边缘使用零值进行填充, 这样我们就可以对输入图像矩阵的边缘进行滤波

## ReLU
* 線性整流單元（Rectified Linear Unit，ReLU）
* 數學原理也很簡單——將圖片上的所有負數轉為 0



* 其他非线性函数，比如 tanh 或者 sigmoid 也可以用来替代 ReLU，但 ReLU 在大部分情况下表现是更好的

## 池化 (Pooling)
* 空间池化（Spatial Pooling）降低了各个特征图的维度，但可以保持大部分重要的信息
* 空间池化有下面几种方式：最大化、平均化、加和等等

* 下面的图展示了使用 2x2 窗口在修正特征图使用最大池化的例子
* 我们以 2 个元素（也叫做“步长”）滑动我们 2x2 的窗口，并在每个区域内取最大值


## 全连接（Fully Connected）
* 全连接层是学习这些特征的非线性组合的简单方法
* 使用这些特征把输入图像基于训练数据集进行分类
* 可以在输出层使用 softmax 作为激活函数进行


## 以可視化例子整理說明

* 输入图像包含 1024 个像素（32 x 32 大小）

* 第一个卷积层（卷积层 1）由六个独特的 5x5 （步长为 1）的滤波器组成  
  => 6 filters(5x5), stride = 1, output = 28x28 matrix
* 后面是池化层 1，在卷积层 1 得到的六个特征图上分别进行 2x2 的最大池化（步长为 2）的操作  
  => max pooling(2x2), stride = 2, output = 14x14 matrix


* 卷积层 2 是十六个 5x5 （步长为 1）的卷积滤波器  
  => 16 filters(5x5), stride = 1, output = 10x10 matrix
* 池化层 2，进行 2x2 的最大池化（步长为 2）的操作
  => max pooling(2x2), stride = 2, output = 5x5 matrix


* 第一个全连接层有 120 个神经元
* 第二层全连接层有 100 个神经元
* 第三个全连接层有 10 个神经元，对应 10 个数字——也就做输出层
