## Introduction
* ILSVR: ImageNet Large Scale Visual Recognition Competition
* 由 ImageNet 所舉辦
* ImageNet 於2007年由史丹福大學教授李飛飛創辦
* 目標是收集大量帶有標註(labeling)信息的圖片數據供計算機視覺模型訓練
* ImageNet 擁有1500萬張標註過的高清圖片，總共擁有22000類，其中約有100萬張標註了圖片中主要物體的定位邊框
* 自 2010 年開辦
* 2012 年深度學習網路 Alexnet, 以Top-5錯誤率15.4%超過10%的懸殊差距，輕取Xerox團隊所使用的SVM機器學習技術
* 自從此屆開始，各種創新的神經網路模型如雨後春荀竄起

## (2012)Alexnet(經典)
0. 8 layers
1. 使用ReLu取代了Sigmoid。
2. 使用Dropout技術。
3. 使用Image augmentation技術。
4. Pooling採用max pooling（取max值）。
5. 使用兩片GTX 580 GPU，針對1,500萬張相片、22,000種類別，運行一個星期訓練完成。

## (2013)ZFnet
* New York University的Matthew Zeiler及Rob Fergus
* ZFnet架構是由Alexnet修改而來
  1. 不同於Alexnet第一層捲積使用11×11的filter，ZFnet改用較小的7×7 filter以保留更多的原始圖片資訊輸入
  2. 隨著層數加深，應用更多數量的filters。例如，從第3至第5層的捲積，Alexnet用了384、384、256個filters，ZFnet則增加到512、1024、512。
  3. ZFnet亦使用Relu作為activate 但Loss function則使用cross-entropy loss。
* 僅使用ImageNet的130萬張圖片，在一片GTX580 GPU上訓練12天完成

## (2014)GoogLeNet(經典)
* 它是由Goog加上LeNet的縮寫而成（為了向LeNet表示敬意）
* 雖然GoogLeNet的層數(22 layers)相當多，但是其運算效率卻相當好，使用的超參數數量硬是比起Alexnet少了12倍，準確率卻反而更高
* GoogLeNet的創新之處在於大量使用了所謂的「Inception」，它是一種network in network的架構
* 針對輸入資料同時併行不同filter尺寸的卷積處理和max pooling，最後再進行級聯，這讓系統在同一層便能取得不同level的特徵

## (2015)ResNet(經典)
* 由微軟研究院所開發
* 全名是Residual Neural Network
* 特點是神經網路不一定要一層層接續往下執行，可以是跳躍方式略過下一層而直達下下層

* Resnet架構的提出主要是為了解決當神經網絡的深度持續增加時所出現的Degradation問題，亦即準確率隨著深度增加後，到了某個深度後會達到飽和無法提昇
* 若再持續增加深度反而會導致準確率下降，其原因不在over-fitting，而是增加training layers反而帶來的training errors

* ResNet其深度是152層，足足是GoogLeNet 22層的七倍，但Top-5 error rate卻大幅降低了47%

## (2016)GBD-Net
* 由中國的CUImage（商湯科技和港中文）提出的GBD-Net拿下
* GBD-Net是基於ResNet-269進行擴充修改

## (2017)SeNet
* 由新加坡國立大學與奇虎360 合作的SeNet拿下
* SeNet全稱是Squeeze-and-Excitation Networks
* 依據loss function的學習來調整不同屬性的特徵權重，讓有效的feature map權重加大，無效或效果小的feature map權重變小，使得模型訓練達到更好的結果
* 這次參與競賽的SeNet便是由ResNet的改進版ResNeXt加上Squeeze-and-Excitation Block的model
