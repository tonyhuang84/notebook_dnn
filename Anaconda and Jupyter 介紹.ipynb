* conda is also a top-notch virtual environment manager

## Setting up a virtual environment for deep learning
- creating a new conda environment with Python  
```
conda create --name deeplearning python
```  
- enter this newly created virtual environment  
```
source activate deeplearning
```



## Install IPython and Jupyter
```
conda install ipython
conda install jupyter
```

## Install libraries for deep learning
```
conda install pandas
conda install scipy
conda install seaborn  (or matplotlib)
conda install scikit-learn
pip install --upgrade tensorflow     (CPU support only)
pip install --upgrade keras
```
